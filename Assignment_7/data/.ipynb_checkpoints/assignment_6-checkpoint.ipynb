```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data\weather.csv')
```

### 1.  A Plot Between Minimum And Maximum Temperature

```python
plt.figure(figsize=(15,10))
df[['MinTemp','MaxTemp','Date']].plot()
```

### 2. A Plot Describing pressure at 9am and 3pm

```python
plt.figure(1)
plt.subplot(121)
sns.distplot(df['Pressure9am'])
plt.subplot(122)
sns.distplot(df['Pressure3pm'])
```

### 3. Plot between Rainfall and Evaporation

```python
sns.lineplot(x="Evaporation",y="Rainfall",data=df)
```

### 4. Plot between Windspeed And Humidity on a Rainy Day at 9am


```python
sns.barplot(x="WindSpeed9am",y="Humidity9am",data=df.iloc[0:50],hue="RainToday")
```

### 5.  Plot Between Pressure and Humidity at 3pm 

```python
sns.lineplot(x='Humidity3pm', y='Pressure3pm', data=df,hue="RainTomorrow")
```

### 6.  Plot Between Pressure and Temperature

```python
sns.jointplot(x="Pressure9am", y="Temp9am",kind='hex',
              data=df)
```

### 7.  Plot Between Pressure and Evaporation

```python
sns.barplot(x="Pressure9am", y="Evaporation",data=df.iloc[0:10],hue='RainTomorrow')
```

### 8.  Plot showing Max Temperature 

```python
PressureDate = df.iloc[0:15,0:15]
sns.barplot(x='MaxTemp',y='Date',data=PressureDate)
```

### 9.  Plot showing Humidity at 9am

```python
sns.distplot(df["Humidity9am"])
```

### 10.  Plot between Sunshine and Evaporation on a Rainy Day

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='Sunshine',y='Pressure9am',data=df,hue="RainToday")
```

### 11.  Wind Gust Speed classification

```python
plt.figure(figsize=(15,8))
df['WindGustDir'].value_counts().plot(kind='pie')
```

### 12.  Plot Between Cloud and Max Temperature

```python
sns.boxplot(x='Cloud9am', y='MaxTemp', data=df,hue="RainToday")
```

### 12.  Plot Between Humidity and Cloud at 9am

```python
sns.jointplot(x = 'Humidity9am', y = "Cloud9am", data= df,kind="kde")
```

### 13.  Plot Between Windspeed and Cloud at 3pm

```python
sns.barplot(x="WindSpeed3pm",y="Cloud3pm" ,data=df)
```

### 14.  Plot for Risk_MM

```python
df["RISK_MM"].plot(kind="kde")
```

### 15.  Plot Between Humidity and Cloud at 9am

```python
sns.barplot(x="Humidity9am",y="Humidity3pm",hue="RainTomorrow",data=df.iloc[0:25])
```

### 16.  Plot between Temperature Pressure Humidity and cloud at 9am on Rainy day

```python
sns.pairplot(data=df[["Temp9am","Humidity9am","Pressure9am","Cloud9am","RainToday"]],hue="RainToday")
```

### 17.  Plot for Pressure at 9am on Rainy day

```python
sns.countplot(y="Pressure9am" ,data=df.iloc[0:25],hue="RainToday")
```

### 18.  Plot for sunshine on Rainy days

```python
sns.barplot(x="Sunshine",y="Date",data=df[:20],hue="RainToday");
```

### 19.  Plot Between Humidity and Cloud at 9am

```python
sns.barplot(x="Humidity9am",y="Cloud9am",data=df[:50])
```

### 20.  Plot for Number of raindy days

```python
sns.countplot(x=df["RainToday"], data=df)
```
